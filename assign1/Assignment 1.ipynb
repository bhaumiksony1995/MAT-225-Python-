```python
#Halfadder function
def halfadder(p,q):
    c = (p and q)
    s = not p and (p or not q)
    return c,s

#Fulladder Function
def fulladder(p,q,r):
    c1,s1 = halfadder(p,q)
    c2,SUM = halfadder(s1,r)
    CARRY = (c1 or c2)
    return CARRY,SUM

#Paralleladder Function
def paralleladd(p,q,r,t,u,v):
    c1,s1 = halfadder(r,v)
    c2,s2 = fulladder(q,u,c1)
    CARRY,s3 = fulladder(p,t,c2)
    return CARRY,s3,s2,s1

print 'p' + '\t' + 'q' + '\t' + 'Carry' + '\t' + 'Sum'
for p in [True,False]:
    for q in [True,False]:
        c,s = halfadder(p,q)
        print p,q,c,s

print 'p' + '\t' + 'q' + '\t' + 'r' + '\t' + 'Carry' + '\t' + 'Sum'
for p in [True,False]:
    for q in [True,False]:
        for r in [True,False]:
            c,s = fulladder(p,q,r)
            print p,q,r,c,s

print 'p' + '\t' + 'q' + '\t' + 'r' + '\t' + 't'+ '\t' + 'u' + '\t' + 'v' + '\t' + 'Carry' + '\t' + 'Sum1' + '\t' + 'Sum2' + '\t' + 'Sum3'
for p in [True,False]:
    for q in [True,False]:
        for r in [True,False]:
            for t in [True,False]:
                for u in [True,False]:
                    for v in [True,False]:
                        c,s3,s2,s1 = paralleladd(p,q,r,t,u,v)
                        print p,q,r,t,u,v,c,s1,s2,s3
```
